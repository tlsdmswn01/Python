## 예외처리
- 오류를 무시하거나 적절히 처리해줘야 할 때 사용.  
- 예외처리는 보다 안정적이고, 오류에 강한 프로그램을 만들 수 있게 도와준다.  
- 예) 자율주행자동차는 사람의 목숨과 밀접한 관련이 있다. 이 프로그램에서 오류가 난다면 자동차는 멈추게 될 것이고 큰 문제로 이어진다. 이를 방지하기 위해 예외처리를 사용하여 오류가 발생해도 프로그램은 멈추지 않지만 어떠한 코드에 오류가 발생했는지 확인할 수 있다.

## 파이썬 오류의 종류  
-> 오류들은 대부분 파이썬 Exception 클래스의 파생 클래스  
-> Exception 클래스를 상속받아서 나만의 오류를 만들 수 있다.    

- ImportError: 임포트 하려는 것을 못 찾음
- ModuleNotFoundError: 모듈을 못 찾음
- SyntaxError: 대부분의 문법 오류
- IndentationError: 들여쓰기 오류
- TypeError: 잘못된 자료형을 사용한 오류
- ValueError: 자료형은 문제없지만 값이 문제인 경우
- ZeroDivisionError: 0으로 나누기를 했을 때
- IndexError: 인덱스가 범위를 벗어난 오류
- KeyError: 참조하려는 키가 기존 키 집합에 없음
- StopIteration: 이터레이터에 의해 생성된 항목이 더 없다는 것을 알려주기 위해 나옴

```python
# 기본문법
try:
    a = [1,2]
    print(a[3])
    4/0
except (ZeroDivisionError, IndexError) as e:
    print(e)
```

```python
#Exception을 사용하면 모든 오류에 대해 처리를 받을 수 있다.
try:
    a = [1,2]
    #print(a[3])
    4/0
except Exception as e:
    print(e)
```

```python
#finally는 오류 상관없이 try문 실행 후 항상 실행된다.
try:
    a = [1,2]
    #print(a[3])
    4/0
except Exception as e:
    print(e)
finally:
    print('okay')
```

```python
#일부러 예외를 발생시켜야 하는 경우
#raise+에러 이름을 붙여 쓰거나 raise 오류 이름('메세지')로도 사용가능
#raise ValueError('값을 확인하세요')
print(1)
a = int(input("1~5 까지 숫자 입력 : "))
if a < 1 or a > 5:
  raise ValueError("밸류 에러") #우리가 원하는 1~5가 아닌 다른 숫자를 사용자가 입력했을 때 오류 발생시키는
print(f"입력한 a : {a} 입니다.")
#raise Error("밸류 에러")
```
