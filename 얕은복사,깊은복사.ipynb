## 얕은 복사, 깊은 복사  
- 얕은 복사는 겉만 스윽 복사해서 주소값이랑 값이 같이 복사되는데,  
깊은 복사는 겉이 아니라 안까지 들어가서 구체적으로 복사하기 때문에, 주소값은 복사하지 않고, 안에 값만 복사한다.


```python
#얕은 복사 -> 주소값만 동일하게
a=[10,9,8,7,6,5,4]
b=[1,2,3,4,5]

print(id(a),id(b))
b=a
print(id(a),id(b))
a.sort()
print(b) #a.sort()에 영향받는다. a에 함수를 적용하면 b에도 적용
```

```python
#깊은 복사 -> 주소값은 다른데 값은 같은것
import copy
a=[5,4,3,2,1]
b=[1,2,3,4,5]

print(id(a),id(b))
b=copy.deepcopy(a)

a.sort()
print(b) # a.sort()에 영향 받지 않는다. a에 함수를 적용해도 깊은 복사는 a와 b가 주소값이 다르기 때문에
# b에는 함수가 적용되지 않는다.
print(a)

#2차원 행렬은 모두 깊은 복사
```
