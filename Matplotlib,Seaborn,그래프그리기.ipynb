### 그래프 기초

```python
import seaborn as sns
ans=sns.load_dataset('anscombe')
print(ans)
print(type(ans))
```

```python
import matplotlib.pyplot as plt

# 데이터 필터링
dataset1=ans[ans['dataset']=='I']
dataset2=ans[ans['dataset']=='II']
dataset3=ans[ans['dataset']=='III']
dataset4=ans[ans['dataset']=='IV']

#그래프 격자 만들기
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)

#그래프 그리기
ax1.plot(dataset1['x'],dataset1['y'],"^")
ax2.plot(dataset2['x'],dataset2['y'],"o")
ax3.plot(dataset3['x'],dataset3['y'],"o")
ax4.plot(dataset4['x'],dataset4['y'],"^")

#각각 그래프 제목 붙이기
ax1.set_title('dataset1')
ax2.set_title('dataset2')
ax3.set_title('dataset3')
ax4.set_title('dataset4')

#그림판 이름 붙이기
fig.suptitle('Ans data')

#그래프 글자 겹치는 부분 해결하기
fig.tight_layout()
```

### 히스토그램

```python
tips=sns.load_dataset('tips')
print(tips.head())
print(type(tips))

#격자 만들기
fig=plt.figure()
al=fig.add_subplot()

# 히스토그램 그래프그리기
al.hist(tips['total_bill'],bins=10) # bins는 간격
al.set_title('Histogram')
al.set_xlabel('Total_bill')
al.set_ylabel('Frequency')
```

### 산점도

```python
#격자 만들기
fig=plt.figure()
al=fig.add_subplot()

#scatter, 산점도 그래프 그리기
al.scatter(tips['total_bill'],tips['tip'])
al.set_title('Scatterplot of TotalBill vs. Tip')
al.set_xlabel('Total_bill')
al.set_ylabel('Tip')
```

### 박스 그래프

```python
#격자 만들기
fig=plt.figure()
al=fig.add_subplot()

#scatter, 산점도 그래프 그리기
al.boxplot([tips[tips['sex']=='Female']['tip'],tips[tips['sex']=='Male']['tip']],labels=['Female','Male'])
al.set_title('Boxplot of Tips by Sex')
al.set_xlabel('Sex')
al.set_ylabel('tip')
```

### 다변량 그래프 그리기

```python
def recode_sex(sex):
    if sex=='Female':
        return 0
    else:
        return 1

tips['sex_color']=tips['sex'].apply(recode_sex) # apply 열에 함수를 적용하는

#격자 만들기
fig=plt.figure()
al=fig.add_subplot()

#scatter, 산점도 그래프 그리기
al.scatter(tips['total_bill'],tips['tip'],tips['size']*10,tips['sex_color'],alpha=0.5) #scatter(x축,y축,점 사이즈, 점 색깔, 불투명도)
al.set_title('Scatterplot of TotalBill vs. Tip Colored by Sex and Sized by Size')
al.set_xlabel('Total_bill')
al.set_ylabel('Tip')
```

## Seaborn 라이브러리로 그래프 그리기

```python
#히스토그램
al=plt.subplots()
al=sns.distplot(tips['total_bill'],kde=False) #kde=False는 히스토그램만 그리기, default는 True
```

```python
al1=plt.subplots()
al1=sns.distplot(tips['total_bill'],hist=False) #his=False 히스토그램을 그리지 않는
```

```python
al2=plt.subplots()
al2=sns.distplot(tips['total_bill'],rug=True) #rug -> 데이터 밀집도 확인 가능
```

```python
ax=plt.subplots()
ax=sns.countplot(x='day',data=tips)
```

```python
# 산점도 그래프 그리기
ax=sns.lmplot(x='total_bill',y='tip',data=tips,fit_reg=True,col='sex',col_wrap=2) # fit_reg =False면 추세선 삭제, col: 어떤 열로 나누어 그래프를 그릴 것인지
#ax.set_title('Scatterplot of Total BIll and TIp')
```

```python
joint= sns.jointplot(x='total_bill',y='tip',data=tips)
joint.set_axis_labels(xlabel='Total_bill',ylabel='Tip')
joint.fig.suptitle('Joint plot of Total Bill and Tip',fontsize=10,y=1.03)
```

```python
joint= sns.jointplot(x='total_bill',y='tip',data=tips, kind='hex')# 육각형 그래프를 그리는, 밀집도를 더 잘 표현
joint.set_axis_labels(xlabel='Total_bill',ylabel='Tip')
joint.fig.suptitle('Joint plot of Total Bill and Tip',fontsize=10,y=1.03)
```

```python
ax=plt.subplots()
ax=sns.kdeplot(x=tips['total_bill'],y=tips['tip'],shade=True)
```

```python
ax=sns.barplot(x='time',y='total_bill',data=tips)
```

```python
ax=sns.boxplot(x='time',y='total_bill',data=tips)
```

```python
ax=sns.violinplot(x='time',y='total_bill',data=tips,hue='day')
```

```python
#관계그래프
fig=sns.pairplot(tips,hue='sex') #자동으로 배치, 단점: 중복되는 그래프가 있음
```

```python
pair_grid= sns.PairGrid(tips) #그림판 만들기
pair_grid=pair_grid.map_upper(sns.regplot) # map_upper : 대각선을 기준으로 위쪽
pair_grid=pair_grid.map_lower(sns.kdeplot,shade=True)
pair_grid=pair_grid.map_diag(sns.distplot,rug=True)
```

### FacetGrid 클래스

```python
facet=sns.FacetGrid(tips, col='day') #tips데이터로 그래프 그리되, day를 기준으로 그래프 나누기
facet.map(sns.distplot,'total_bill',rug=True) # map(원하는 그래프,x축 rug: 밀집도 확인)
```

```python
facet=sns.FacetGrid(tips, col='day',hue='sex') #tips데이터로 그래프 그리되, day를 기준으로 그래프 나누기
facet.map(plt.scatter,'total_bill','tip')
```

```python
facet=sns.FacetGrid(tips, col='time',row='smoker',hue='sex',) #time: launch,dinner / smoker: Yes,No
facet.map(plt.scatter,'total_bill','tip')
```
