## 파일 불러오고 저장하기
### 파일 불러오기
- csv 파일 불러오기 : pd.read_csv()  
- tsv 파일 불러오기 : pd.read_tsv('파일명.tsv',sep='\t')  

- 엑셀파일 불러오기 : pd.read_excel('파일이름.xlsx',sheet_name=시트이름)  
- Picke파일 불러오기 : pd.read_pickle('파일이름.pickle')
- 매개변수
 encoding(문자표현방식) -> cp949,euc-kr    
 sep(데이터를 읽어올 때 구분자)->\t(tsv파일)
### 파일 저장하기
- csv 파일 저장하기 : df.to_csv()  
- tsv 파일 저장하기 : df.read_tsv('파일명.tsv',sep='\t')  
- 엑셀파일 저장하기 : df.to_excel('파일이름.xlsx',sheet_name=시트이름)  
- Picke파일 불러오기 : df.to_pickle('파일이름.pickle')
- 매개변수
index=True/False: 인덱스 유무

- Pickle 파일의 장점  
pickle로 데이터를 저장하면 더 많은 데이터를 더 적은 용량에 저장이 가능, 속도도 다른 파일보다 더 빠른 속도를 가진다.

## 시리즈(Series)  
- 데이터프레임에서 각각의 열을 시리즈라고 한다.  
- 판다스의 pd.Series()함수를 사용해 시리즈 생

```python
#딕셔너리를 사용해 시리즈 만들기
dic={'a':1,'b':2,'c':3} #인덱스가 key값이 되는
pd.Series(dic)
#리스트를 사용해 시리즈 만들기
box=[1,2,3]
s=pd.Series(box,index=['a','b','c'])
type(s)
s.index #시리즈 행 이름 확인
s.values #시리즈 값 확인
```

```python
#예제
data=[1,2,3,4,5,6,7,8]
data1=['A','B','C','D','F','G','H','I']
series=pd.Series(data, index=data1) # 원하는 인덱스 명으로 시리즈 만들기
series.index # 인덱스 추출
series.values # series 값 추출

series=series*100
series1=series[series > 500]
print(series)
print(series1)
```

## 유용한 데이터프레임 함수
### Group by
- 판다스에서 제공하는 데이터프레임을 집단별로 나누어 주는 함수, 각 집단별로 통계값을 구할 때 유용하게 사용 가

```python
# 각 출생년도 별로 기대수명 평균 구하기
df=pd.read_csv('/content/drive/MyDrive/Github/Python/실습 데이터_데이터 분석 기초/gapminder.tsv','\t')
result=df.groupby('year')['lifeExp'].mean()
#result.plot()
result2=df.groupby(['year','continent'])['lifeExp'].nunique() #n: 특정한 것을 세다
result2
```
