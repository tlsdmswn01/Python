## 날짜와 시간 모듈 Datetime  
- 날짜와 시간에 관한 기능을 제공하고 조작할 수 있는 모듈  
- 아래의 datetime.datetime.now()에서 앞의 datetime은 모듈의 이름, 뒤의 datetime은
클래스의 이름

```python
from datetime import datetime as dd # datetime 모듈에서 datetime 클래스를 불러오고 별명을 dd로 하겠다.
#dd.datetime.now() #현재 년,월,일,시간을 알려줌
today=dd.today() #datetime 모듈에서, date 클래스에서 today함수 사용
print(today)
today.year #현재 년도만 가져오는
today.month
today.day
today.day
```

```python
# 모듈이 가진 클래스의 목록을 출력 - dir함수
dir(dd)
```

```python
# Replace메소드 - datetime에서 날짜나 시간 값을 변경하고 싶을 때 사용
start_time= dd.datetime.now()
start_time.replace(month=12,day=25)
```

```python
# datetime 모듈을 사용해 크리스마스까지 남은 시간 구하기
print('오늘은 {}년 {}월 {}일입니다.'.format(today.year,today.month,today.day))
Xmas= dd(2023,12,25)
time_gap= Xmas - dd.now()
print('다음 크리스마스 까지는 {}일 {}시간 남았습니다'.format(time_gap.days,time_gap.seconds//3600))
```

```python
# 100일 뒤 날짜 구하기
print('오늘 =', dd.now())
hundred=dd.timedelta(days=100) # 100일 경과시간
plus100day=dd.now()+hundred # 현재 시간에서 100일 경과시간을 더함
print('100일 후 =', plus100day())
```

## Time 모듈  
- 시간에 관련된 정보 제공  
- 유닉스 시스템의 시작 시간인 1970년 1월 1일 0시 0분 0초 협정 세계시(UTC)를 에폭
epoch이라고 함  
-  유닉스 운영체제에서 표준으로 사용되는 시간 체계는 에폭 시간 혹은 유닉스 시간이라고도
한다.


```python
import time
seconds = time.time()
print('에폭 이후의 시간 = ', seconds)
```

```python
#strftime() - localtime()함수가 반환한 struct_time이라는 형식의 튜플 값을 지정된 형태의 문자열로 반환
import time
unix_timestamp = time.time()
local_time = time.localtime(unix_timestamp)
print(time.strftime('%Y-%m-%d %H:%M:%S', local_time))

#thread - 한 프로그램 안에서 실행되는 작은 실행단위
#sleep() - 일정한 시간동안 현재 실행중인 thread를 일시 중지
time.sleep(1.5) # 1.5초 후 실행
start_time=time.time()
print(1+2+3+4+5)
end_time= time.time()
gap=end_time-start_time
print('1에서 5까지의 합을 구하고 출력하는 시간:{:7.4f}'.format(gap))
```

## 랜덤 모듈 Random  
- 임의의 수를 생성하거나, 리스트 내의 원소를 무작위로 섞거나 선택하는 함수를 포함
image.png

```python
import random as rd
rd.random()
rd.randrange(1,7) # 지정된 범위 내의 정수를 반환, 1이상 7미만, range 범위 내의 값들 중 랜덤으로 반환
rd.randint(1,10) # 1이상 10이하의(1,10이 포함)임의의 정수를 반환
a=[1,2,3,4]
rd.shuffle(a) # 시퀀스의 원소를 랜덤하게 섞어 반환
print(a)
rd.choice(a)#인자로 들어온 시퀀스로부터 임의의 원소를 추출
rd.sample(a,3) #반환할 원소의 개수를 인자로 넣어줄 수 있다.
```

## 의사랜덤과 시드번호  
- 많은 프로그래밍 언어에서 사용하는 랜덤 함수는 **의사랜덤** 생성기로 만들어 낸다.  
- X는 의사 랜덤 값의 열(sequence)가 되며, Xn값은 Xn+1 값을 만드는데 사용될 수 있다.

```python
#의사 랜덤 함수를 이용해 난수 만들기
def pseudo_rand(x):
    a=110354654
    b=12345
    m=2**31
    new_x=(a*x+b)%m
    return new_x
x=pseudo_rand(100) # seed number (씨앗 값)
print(x)
x=pseudo_rand(101)
print(x)

for _ in range(5):
    x=pseudo_rand(x)
    print(x)
```

```python
#로또 번호 만들기
# 로또는 1에서 45 사이의 임의의 정수를 6개 맞히는 규칙이 있다
num=list(range(1,46))
rd.shuffle(num)
num1=num[:7]
num1.sort()
print(num1)
```
