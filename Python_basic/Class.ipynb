#Class  
- 여러 개의 객체를 만들어내는 기능  
- 예) list,tuple,dic 등 각 자료형들 또한 하나의 객체. list의 경우 list만 사용할 수 있는 함수들이 있고, dic또한 해당 자료형에서만 사용할 수 있는 함수가 따로 있다. 즉 class로 객체를 생성하고, 각 객체들마다 사용할 수 있는 함수들이 다르다는 것을 알 수 있다.

```python
#기본문법
class Bank:
    def __init__(self):
        self.money=0

    def deposit(self,a):
        self.money+=a

    def withdraw(self,a):
        self.money-=a

a=Bank()
a.deposit(5000)
a.withdraw(4000)
print(a.money)
```
