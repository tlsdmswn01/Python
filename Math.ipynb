### 연립방정식

```python
import numpy as np
# 5x-(-2)y=-13
# 4x-5y=-6
a=np.array([[5,-2],[4,5]])
b=np.array([-13,-6])
x=np.linalg.solve(a,b)# 연립방정식 해 구하기 (x,y)
x
```

### Sympy를 활용한 미분

```python
import sympy
x=sympy.symbols('x')
f=x*sympy.exp(x) # y=x*e^x
print(f)
f_diff=sympy.diff(f)
print(f_diff)
```

### 확률과 통계

```python
import numpy as np
# 평균
#np.mean()
#중앙값
#np.median()
#최빈값
#np.mode()
#편차 - 평균으로부터 얼마나 떨어져 있는가
scores=np.array([1,2,3,4,5,6,10,20,40,10,20,2,1,4,3])
mean=np.mean(scores)
deviation=scores-mean
deviation
#분산
np.var(scores)
#표준편차
np.sqrt(np.var(scores))
```

### 데이터의 정규화
시험점수가 동일하더라도, 쉬운 시험과 어려운 시험에서의 60점은 상대적인 결과가 다르다. 점수는 평균과 분산에 따라 평가가 달라진다.   
데이터를 통일된 지표로 변환하는 것을 정규화라고 한다.

```python
# 데이터의 표준화 - 평균이 0, 표준편차가 1
z= (scores-np.mean(scores))/np.std(scores)
z
```

### 상관계수
- 상관계수는 -1~1사이의 값을 가진다.  
- 1또는 -1에 가까워질수록 높은 상관관계가 있다  
- 0에 가까울수록 서로 상관관계가 없다.

```python
import pandas as pd
col1 = [1,2,3,4,5,6]
col2 = [1,4,2,8,16,32]
col3 = [6,5,4,3,2,1]
data = {"col1":col1,"col2":col2,"col3":col3}
df = pd.DataFrame(data)
print(df)
corr=df.corr(method='pearson') # 변수들간의 상관계수
column_names=['col1','col2','col3']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# 산점도 작성
plt.figure(figsize=(10, 10))         # 플롯 사이즈 지정
plt.scatter(df.col1, df.col2) # 산점도
plt.xlabel("Sales")                  # x축 레이블 지정
plt.ylabel("GDP1")                   # y축 레이블 지정
plt.grid()                           # 플롯에 격자 보이기
plt.show()                           # 플롯 보이기

import seaborn as sns

# 레이블의 폰트 사이즈를 조정
sns.set(font_scale=1.5)
hm = sns.heatmap(corr.values,
            cbar=True,
            annot=True,
            square=True,
            fmt='.2f',
            annot_kws={'size': 15},
            yticklabels=column_names,
            xticklabels=column_names)

plt.tight_layout()
plt.show()
```
