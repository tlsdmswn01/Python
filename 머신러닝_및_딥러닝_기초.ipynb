## 지도학습에 필요한 데이터 가공하기  
- 지도학습: 문제와 답을 같이 학습함으로써 미지의 문제에 대한 올바른 답을 예측하고 하는 방법  
- 문제집 : 2차원 이상의 행렬  
- 정답지 : 1차원 행렬  
- 데이터 프레임이 아닌 행렬 형태인 이유  
-> 병렬계산을 하기 위해서

```python
import pandas as pd
fish=pd.read_csv('/content/drive/MyDrive/Github/Python/실습 데이터_데이터 분석 기초/fish.csv')
fish # class 1:도미, 0:방어
```

```python
bream=fish[fish['class']==1] # 도미
smelt=fish[fish['class']==0] # 방어

bream_length=list(bream['length'])
bream_weight=list(bream['weight'])

smelt_length=list(smelt['length'])
smelt_weight=list(smelt['weight'])
```

```python
print(bream_length) # 대부분 길이가 20을 넘어감
print(smelt_length)
```

```python
import matplotlib.pyplot as plt
plt.scatter(bream_length,bream_weight)
```

```python
plt.scatter(smelt_length,smelt_weight)
```

```python
plt.scatter(bream_length,bream_weight)
plt.scatter(smelt_length,smelt_weight)
plt.legend(['bream','smelt'])
```

```python
# 문제집(2차원 행렬) : 길이, 무게 - 컴퓨터가 길이와 무게만 보고 어떤 물고기가 맞추는
# 정답지(1차원 행렬) : 도미, 방어(생선종류)
# 도미: 35마리, 방어: 14마리
length = bream_length + smelt_length
print(length)
weight= bream_weight + smelt_weight
print(weight)
```

```python
# 문제지
data=[]

for i in range(49):
    data.append([length[i],weight[i]])

data # 2차원 리스트
```

```python
# 정답지
target=[1]*35+[0]*14
print(target)
```

```python
# 행렬로 변환
import numpy as np
data=np.array(data)
data
target=np.array(target)
target
```

- 장점 : 간단한 원리, 우수한 성능, 이진분류, 다중분류 모두 가능

## 데이터를 분할해서 학습  
- 훈련데이터와 시험데이터  
훈련데이터로 학습하고, 훈련데이터로 평가한다면 정확도 신뢰할 수 없음  
훈련데이터와 시험데이터를 만들고, 훈련데이터로만 학습하고, 시험데이터로 평가  
- 데이터의 비율  
최대한 훈련데이터에 많은 비율을 할당하는 것이 좋음  
1. 데이터가 적음 -> 75:25 비율을 유지  
2. 데이터가 많음(ex. 100만개 이상) -> 99:1 또는 99.9:0.1 수준  

+ 랜덤으로 데이터를 분할하는 것은 비추천  
8:2로 데이터 분할시 훈련데이터에만 도미가 몰려있을 수도 -> 일반적으로 훈련데이터에서 도미와 방어의 비율이 6:4라면 시험데이터에서도 도미와 방어의 비율이 6:4인 것이 좋다  
이를 해결해주는 모듈이 train_test_split

```python
from sklearn.model_selection import train_test_split
#훈련문제집, 시험문제집, 훈련정답지, 시험정답지 = train_test_split(문제집, 정답지, test_size=0.2)
#test_size: 시험 데이터 할당정도, 데이터의 20%를 시험데이터로 할당할 것이다.

```

```python
#실습
train_input, test_input, train_target, test_target = train_test_split(data, target, test_size=0.2) # test_size default=0.25
knn1=KNeighborsClassifier() #모델생성
knn1.fit(train_input, train_target)
```

```python
knn1.score(test_input,test_target)
test_target
```

```python
knn1.predict(test_input)
```
